```python
columns = ['customer_age' , 'quantity', 'price']
```

```python
columns = ['customer_age' , 'quantity', 'price']

for col in columns:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)

    IQR = Q3-Q1

    LB = Q1-1.5*IQR
    UB = Q3+1.5*IQR

    md = df[col].median()

    outlier = (df[col] < LB) | (df[col] > UB)
    
outlier.sum()
```

```python
columns = ['customer_age' , 'quantity', 'price']

for col in columns:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)

    IQR = Q3-Q1

    LB = Q1-1.5*IQR
    UB = Q3+1.5*IQR

    df[col] = np.where(df[col] < LB, LB, np.where(df[col] > UB, UB, df[col]))
              

    

```

```python
plt.figure(figsize=(8,4))
sns.boxplot(data=df[columns])
plt.title("Boxplots of Selected Columns")
plt.show()
```

#  Trend Analysis

```python
df['Year'] = df['purchase_date'].dt.year
df['Month'] = df['purchase_date'].dt.to_period('M')
df['Quarter'] = df['purchase_date'].dt.to_period('Q')
```

```python
plt.figure(figsize = (15,4))

sales_by_day = df.groupby('purchase_date')['Total_Sales'].sum().plot(kind = 'line')
plt.title("Day-wise Sales")
```

```python
plt.figure(figsize = (15,4))
sales_by_month = df.groupby('Month')['Total_Sales'].sum().plot(kind = 'line')
plt.title("Month-wise Sales")
```

```python
plt.figure(figsize = (15,4))
sales_by_quarter = df.groupby('Quarter')['Total_Sales'].sum().plot(kind = 'line')
plt.title("Quarter-wise Sales")
```

```python

plt.figure(figsize = (15,4))
sales_by_year = df.groupby('Year')['Total_Sales'].sum().plot(kind = 'line')
plt.title("Year-wise Sales")
```

Sales are decreasing across all time periods (daily, monthly, quarterly, yearly).
